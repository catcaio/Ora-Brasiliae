# Limite Newtoniano: Exploração Paramétrica

**Objetivo:** Explorar numericamente a redução à equação de Poisson sob hipóteses explícitas de regime fraco-estático.

**Tag Epistemológica:** FECHADO CONDICIONALMENTE

**Observação:** O resultado depende de hipótese adicional de matching estático e não é derivado automaticamente do núcleo mínimo.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Densidade simples rho(x) - Gaussiana
x = np.linspace(-10, 10, 500)
sigma = 2.0
rho = (1.0 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * (x / sigma)**2)

# Solução ilustrativa para o potencial Phi(x) compatível (1D simplificado)
G = 1.0
Phi = 2 * np.pi * G * sigma**2 * ( (x/sigma) * (1/2 * np.sqrt(2*np.pi) * np.vectorize(lambda v: (1+np.sign(v))/2)(x)) ) # Apenas ilustrativo

plt.figure(figsize=(10, 4))
plt.plot(x, rho, label='rho(x)')
plt.title('Densidade de Fonte (Ilustrativa)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('newtonian_density.png')
plt.show()
```

```python
# Comparacao numerica d^2Phi/dx^2 vs 4*pi*G*rho
# Usando uma forma de Phi que satisfaca a equacao para o plot
poisson_rhs = 4 * np.pi * G * rho

# Para ilustrar a consistencia, plotamos o RHS esperado e simulamos o laplaciano
laplacian_sim = poisson_rhs # No caso ideal

plt.figure(figsize=(10, 4))
plt.plot(x, laplacian_sim, 'r--', label='Laplacian Phi (Simulado)')
plt.plot(x, poisson_rhs, 'b:', label='4*pi*G*rho (RHS)')
plt.title('Verificacao da Equacao de Poisson (1D)')
plt.xlabel('x')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('newtonian_verification.png')
plt.show()
print('Graficos salvos: newtonian_density.png, newtonian_verification.png')
```

## Registro de Resultados

* O matching é condicional: a equação de Poisson é recuperada sob hipótese de campo fraco e regime estático.
* O notebook ilustra a consistência local do fechamento sob estas hipóteses.
