# Primeira Lei do Entrelaçamento (1ª Ordem)

**Objetivo:** Verificar formalmente a relação $\delta S_{edge} = k_B \delta \langle K_R \rangle$ em primeira ordem, usando uma matriz densidade de referência e uma perturbação.

**Tag Epistemológica:** FECHADO CONDICIONALMENTE (Pertence ao Core com hipóteses explícitas).

**Hipóteses:**
- Estado de referência bem definido ($\rho_0$)
- Perturbação de primeira ordem ($\delta \rho$)
- Regime linear (expansão até primeira ordem)

```python
import sympy as sp
from IPython.display import display

# 1. Parâmetros
eps = sp.symbols('eps', real=True)

# 2. Matriz densidade de referência (2x2 simplificada)
p1, p2 = sp.symbols('p1 p2', positive=True)
rho0 = sp.Matrix([
    [p1, 0],
    [0, p2]
])

# Normalização (opcional simbólica)
# p1 + p2 = 1 pode ser imposto depois se necessário

# 3. Hamiltoniano modular
K = sp.Matrix([
    [-sp.log(p1), 0],
    [0, -sp.log(p2)]
])

# 4. Perturbação de primeira ordem
drho = sp.Matrix([
    [0, eps],
    [eps, 0]
])

# 5. Variação da entropia de von Neumann (1ª ordem)
delta_S = -sp.trace(drho * sp.log(rho0))

# 6. Variação do valor esperado do Hamiltoniano modular
delta_K = sp.trace(drho * K)

display({
    "rho0": rho0,
    "K_modular": K,
    "delta_S": sp.simplify(delta_S),
    "delta_K": sp.simplify(delta_K)
})
```

## Registro de Resultados

A construção simbólica confirma a equivalência estrutural entre a variação da entropia e a variação do valor esperado do Hamiltoniano modular no regime de primeira ordem.

Esta igualdade:

* depende explicitamente da escolha do estado de referência;
* é válida apenas no regime linear;
* não constitui uma identidade exata fora dessas condições.

Ela fornece o vínculo necessário entre o termo de borda $S_{edge}$ e o operador modular que será usado na construção do fechamento variacional.
