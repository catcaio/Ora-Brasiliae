# Domínio Causal e Definições Locais

**Objetivo:** Definir simbolicamente o domínio geométrico (diamante causal local), a base nula e a métrica no regime local.

**Tag Epistemológica:** DERIVADO

**Definição:** Um diamante causal local é caracterizado por uma superfície de bifurcação $\Sigma$, onde definimos uma base nula com vetores $k$ e $L$ tais que seu produto interno seja constante.

```python
import sympy as sp

G, hbar, c, k_B = sp.symbols('G hbar c k_B', positive=True, real=True)

eta = sp.diag(-1, 1, 1, 1)

k = sp.Matrix([1, 1, 0, 0])
L = sp.Matrix([sp.Rational(1, 2), -sp.Rational(1, 2), 0, 0])

A, S_edge = sp.symbols('A S_edge', real=True)
S_gen = A / (4 * G * hbar) + S_edge

k_null = (k.T * eta * k)[0]
L_null = (L.T * eta * L)[0]
k_dot_L = (k.T * eta * L)[0]

display({
    "eta": eta,
    "k": k,
    "L": L,
    "S_gen": S_gen,
    "k_null": sp.simplify(k_null),
    "L_null": sp.simplify(L_null),
    "k_dot_L": sp.simplify(k_dot_L),
})
```

## Registro de Resultados

Conforme as definições explícitas acima, temos que:
- `eta` = diag(-1, 1, 1, 1)
- `k` é nulo ($k_{null} = 0$)
- `L` é nulo ($L_{null} = 0$)
- `k \cdot L = -1`
- `S_{gen} = A / (4G\hbar) + S_{edge}`
