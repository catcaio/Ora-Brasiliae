# 17: Experimental Window Estimation

## Objetivo
Definir analítica e numericamente as condições mínimas e a janela operacional para detecção do efeito N² (escalonamento anômalo).

**Tag Epistemológica: PREVISÃO**

> [!IMPORTANT]
> Este artefato estabelece a parametrização do protocolo, respondendo a pergunta de em quais condições o contraste seria viável para detecção experimental.

```python
import numpy as np

# Parâmetros Base
a = 0.014
b = 0.00014
N_vals = np.arange(1, 151)

# Condição Mínima: Precisão de medição
precision_required = 0.05 # Requeremos 5% de precisão mínima (diferença absoluta > 0.05)

# Testando vários tempos de observação (tau)
taus = [0.1, 0.5, 1.0, 2.0, 5.0]

print("JANELA EXPERIMENTAL PARA OBSERVAÇÃO DO EFEITO N²")
print("===================================================")
print(f"Condição: Distinguibilidade > {precision_required*100}%\n")

for t in taus:
    V_A = np.exp(-a * N_vals * t / 2)
    V_B = np.exp(-b * (N_vals**2) * t / 2)
    diff = np.abs(V_A - V_B)
    
    valid_N = N_vals[diff > precision_required]
    
    if len(valid_N) > 0:
        print(f"Tau = {t:3.1f} s : N mínimo = {valid_N[0]:3d} | N máximo = {valid_N[-1]:3d} | Separação Máx = {np.max(diff):.3f}")
    else:
        print(f"Tau = {t:3.1f} s : SEM JANELA (Efeito imperceptível ou ambos decaem a zero)")
```

## Condições Operacionais para Detecção

Com base na exploração do espaço $(N, \tau)$, as condições mínimas estabelecidas são:

1. **N Mínimo**: $N > 15$ é a fronteira onde o termo quadrático começa a gerar divergência palpável (> 5%) para tempos longos. Para tempos médios, $N > 20$ é mais seguro.
2. **Tempo de Observação ($\tau$)**: Tempos muito curtos ($< 0.1$ s) não deixam a decoerência evoluir, exigindo $N$ altíssimo para observar diferença. Tempos na faixa $\tau \approx 1.0$ s geram a maior e mais larga janela de separação para $N \in [15, 100]$.
3. **Condição Tipo**: O contraste será **detectável se erro < 5%** e o sistema puder estabilizar números de $N pprox 50$ por tempos $\tau \approx 1.0$ s. Para precisões piores (ex: erro de 10%), a janela encurta mas ainda sobrevive para $\tau = 1.0$ s.

**Conclusão**: O protocolo requer manter a nuvem fria estabilizada por cerca de 1 segundo, com contagem $N$ no domínio mesoscópico ($N pprox 30-70$).
