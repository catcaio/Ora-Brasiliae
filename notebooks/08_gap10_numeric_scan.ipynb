# GAP-10: Varredura Numérica

**Objetivo:** Ilustrar numericamente a dependência logarítmica robusta e a variação do prefator com o parâmetro de esquema.

**Tag Epistemológica:** FECHADO CONDICIONALMENTE

**Observação:** Este notebook NÃO fecha universalidade do coeficiente. Ele apenas ilustra o comportamento numérico da forma logarítmica sob diferentes valores de `scheme_factor`.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# grade para razao wUV/wIR
ratio = np.linspace(1.01, 100, 500)
alpha = 1.0

# forma logaritmica robusta
I_log = alpha * np.log(ratio)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(ratio, I_log, 'k-', linewidth=2)
ax.set_xlabel('wUV / wIR')
ax.set_ylabel('I_log = alpha * log(wUV / wIR)')
ax.set_title('Estrutura logaritmica robusta (alpha = 1)')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('gap10_log_robust.png', dpi=100)
plt.show()
print('Grafico salvo: gap10_log_robust.png')
```

```python
# familias parametrizadas por scheme_factor
scheme_factors = [0.5, 1.0, 1.5, 2.0, 2.5]

fig, ax = plt.subplots(figsize=(8, 5))
for sf in scheme_factors:
    I_scheme = sf * alpha * np.log(ratio)
    ax.plot(ratio, I_scheme, linewidth=1.5, label=f'scheme_factor = {sf}')

ax.set_xlabel('wUV / wIR')
ax.set_ylabel('I_scheme = scheme_factor * alpha * log(wUV / wIR)')
ax.set_title('Dependencia do prefator no parametro de esquema')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('gap10_scheme_families.png', dpi=100)
plt.show()
print('Grafico salvo: gap10_scheme_families.png')
```

## Registro de Resultados

* A estrutura logarítmica é robusta: cresce monotonicamente com a razão wUV/wIR.
* O prefator segue dependente de esquema: famílias distintas de `scheme_factor` geram curvas distintas.
* Este notebook continua fora do core e não fecha universalidade do coeficiente.
