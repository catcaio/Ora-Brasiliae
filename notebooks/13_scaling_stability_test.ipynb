# 13: Scaling Stability Test

## Objetivo
Testar a robustez das assinaturas de escalonamento ($N$ vs $N^2$) sob condições de ruído experimental simulado (ruído em $\gamma$ e jitter em $\tau$).

**Tag Epistemológica: PREVISÃO**

```python
import numpy as np
import matplotlib.pyplot as plt

SEED = 20260423
rng = np.random.default_rng(SEED)

def visibility(tau, gamma):
    return np.exp(-gamma * tau / 2)

# Parâmetros base
a = 0.05
b = 0.005
tau_ref = 1.0
N_range = np.arange(1, 101)
n_simulations = 50
noise_level = 0.1  # 10% de ruído relativo em gamma
jitter_level = 0.05 # 5% de jitter em tau

results_A = []
results_B = []

for _ in range(n_simulations):
    # Jitter em tau
    tau_sim = tau_ref * (1 + rng.normal(0, jitter_level, len(N_range)))
    
    # Ruído em gamma
    gamma_A = (a * N_range) * (1 + rng.normal(0, noise_level, len(N_range)))
    gamma_B = (b * N_range**2) * (1 + rng.normal(0, noise_level, len(N_range)))
    
    results_A.append(visibility(tau_sim, gamma_A))
    results_B.append(visibility(tau_sim, gamma_B))

avg_A = np.mean(results_A, axis=0)
std_A = np.std(results_A, axis=0)
avg_B = np.mean(results_B, axis=0)
std_B = np.std(results_B, axis=0)

plt.figure(figsize=(10, 6))
plt.plot(N_range, avg_A, 'b-', label='Regime A (Avg)')
plt.fill_between(N_range, avg_A - std_A, avg_A + std_A, color='b', alpha=0.2)

plt.plot(N_range, avg_B, 'r-', label='Regime B (Avg)')
plt.fill_between(N_range, avg_B - std_B, avg_B + std_B, color='r', alpha=0.2)

plt.title("Estabilidade do Escalonamento sob Ruído (10% Gamma, 5% Jitter Tau)")
plt.xlabel("N")
plt.ylabel("Visibilidade")
plt.legend()
plt.grid(True)
plt.show()
```

## Análise de Robustez

Observamos que, mesmo com ruído significativo:
1. A separação entre as curvas azul (linear) e vermelha (quadrática) permanece clara para $N > 40$.
2. A variância aumenta com $N$ no regime linear, mas no regime quadrático a curva decai tão rápido que a variância absoluta diminui (embora a relativa possa ser alta).
3. O regime quadrático é altamente resiliente como assinatura qualitativa.
