# Bianchi e φ(x)

**Objetivo:** Documentar simbolicamente a extensão covariante em que um campo efetivo φ(x) entra como completamento consistente, sem promovê-lo a derivação inevitável do núcleo.

**Tag Epistemológica:** COMPLETAMENTO CONSISTENTE

**Escopo:** Este notebook apenas registra a forma simbólica do termo corretivo necessário para compatibilizar a extensão com a identidade de Bianchi.

```python
import sympy as sp
from IPython.display import display

x, y, z, t = sp.symbols('x y z t', real=True)
phi = sp.Function('phi')(t, x, y, z)

G, c = sp.symbols('G c', positive=True, real=True)

# forma simbólica do termo corretivo
J_info = sp.Symbol('J_info_mu_nu')

display({
    "phi_field": phi,
    "corrective_current_symbol": J_info,
    "status": "ansatz_covariant_extension"
})
```

## Registro de Resultados

A execução fixa apenas a estrutura simbólica mínima da extensão:

* φ(x) entra como ansatz de covariância;
* J_info permanece como termo corretivo formal;
* não há demonstração de inevitabilidade do setor escalar-tensorial.
