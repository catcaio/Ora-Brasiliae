# Operador Variacional Projetado

**Objetivo:** Construir simbolicamente o operador variacional $d^2S_{gen}/d\lambda^2$ e o projetor nulo-local $P^{\mu\nu} = k^\mu k^\nu$, que governam a evolução do funcional de entropia ao longo das deformações do horizonte.

**Tag Epistemológica:** DERIVADO (Pertence ao Core).

**Definição:** A condição de estacionariedade local exige a avaliação da Hessiana funcional projetada. No regime nulo-local, a projeção relevante contrai com as direções dos geradores do horizonte.

```python
import sympy as sp
from IPython.display import display

# 1. Constantes e parâmetro afim
G, hbar = sp.symbols('G hbar', positive=True, real=True)
lam = sp.symbols('lam', real=True)

# 2. Funcionais dependentes do parâmetro afim
A = sp.Function('A')(lam)
S_edge = sp.Function('S_edge')(lam)

# 3. Definição de S_gen ao longo da deformação
S_gen = A / (4 * G * hbar) + S_edge

# 4. Operador variacional mínimo
d2Sgen_dlam2 = sp.diff(S_gen, lam, 2)

# 5. Vetor nulo local e projetor nulo-local
k = sp.Matrix([1, 1, 0, 0])
P_mu_nu = k * k.T

display({
    "S_gen(lam)": S_gen,
    "d2Sgen_dlam2": d2Sgen_dlam2,
    "P_mu_nu": P_mu_nu
})
```
