# 14: Model Discrimination Metric

## Objetivo
Quantificar a distinguibilidade entre o regime linear (A) e o quadrático (B) através de uma métrica de distância em função de $N$.

**Tag Epistemológica: PREVISÃO**

```python
import numpy as np
import matplotlib.pyplot as plt

def visibility(tau, gamma):
    return np.exp(-gamma * tau / 2)

a = 0.05
b = 0.005
tau_ref = 1.0
N_vals = np.arange(1, 151)

V_A = visibility(tau_ref, a * N_vals)
V_B = visibility(tau_ref, b * N_vals**2)

# Métrica de Distinguibilidade D(N)
D_N = np.abs(V_A - V_B)

plt.figure(figsize=(10, 6))
plt.plot(N_vals, D_N, 'k-', label='Distinguibilidade $|V_A - V_B|$')
plt.fill_between(N_vals, 0, D_N, color='gray', alpha=0.3)

plt.axhline(y=0.2, color='r', linestyle='--', label='Limiar de Detecção (20%)')

n_max = N_vals[np.argmax(D_N)]
d_max = np.max(D_N)
plt.plot(n_max, d_max, 'ro')
plt.annotate(f'Máximo em N={n_max}', (n_max, d_max), textcoords="offset points", xytext=(0,10), ha='center')

plt.title("Métrica de Distinguibilidade vs N")
plt.xlabel("N")
plt.ylabel("|V_A - V_B|")
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusões sobre a Separação de Modelos

1. **Região de Separação Clara:** Para os parâmetros adotados, os modelos se separam claramente (distância > 0.2) na faixa de $N \in [20, 100]$.
2. **Ponto Ótimo de Medição:** A maior sensibilidade para distinguir os regimes ocorre em $N \approx 50$ para estes parâmetros.
3. **Convergência Tardia:** Para $N$ muito grandes ($> 150$), ambos os modelos convergem para visibilidade zero, reduzindo a distinguibilidade novamente.
