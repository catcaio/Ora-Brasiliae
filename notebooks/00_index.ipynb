# Teoria de Ora Brasiliae: Auditoria Simbólica

## Escopo do Repositório
Este diretório contém a infraestrutura de auditoria simbólica e numérica reprodutível do programa local baseado em entropia generalizada em diamantes causais.

## Fluxo Lógico
`core` -> `gaps` -> `appendices`

## Convenções de Notação
- Métricas e tensores no regime local.
- `k` e `L` como vetores nulos conjugados.
- Assinatura da métrica: `(-, +, +, +)`.
