# Limite Newtoniano

**Objetivo:** Registrar simbolicamente a redução ao limite de Poisson sob hipóteses explícitas de campo fraco e regime estático.

**Tag Epistemológica:** FECHADO CONDICIONALMENTE

**Escopo:** Este notebook não promove o limite newtoniano ao núcleo mínimo; ele apenas fixa a forma simbólica do fechamento sob hipóteses adicionais.

```python
import sympy as sp
from IPython.display import display

x, y, z = sp.symbols('x y z', real=True)
G = sp.symbols('G', positive=True, real=True)

Phi = sp.Function('Phi')(x, y, z)
rho = sp.Function('rho')(x, y, z)

laplacian_Phi = sp.diff(Phi, x, 2) + sp.diff(Phi, y, 2) + sp.diff(Phi, z, 2)
poisson_rhs = 4 * sp.pi * G * rho

display({
    "laplacian_Phi": laplacian_Phi,
    "poisson_rhs": poisson_rhs,
    "poisson_equation_form": sp.Eq(laplacian_Phi, poisson_rhs)
})
```

## Registro de Resultados

A execução fixa a forma simbólica do limite:

* ∇²Φ = 4πGρ
* o resultado depende de hipóteses de campo fraco;
* o resultado depende de matching estático ou quase-estático;
* não pertence ao claim mínimo.
